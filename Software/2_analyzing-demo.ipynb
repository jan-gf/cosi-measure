# Convert to Magnet Coords and Plot Center Slices

```python
from postprocessing.utilities.convert_to_magnet_coords_OSI2OneV1 import convert_to_magnet_coordinate_frame
from postprocessing.visualizer.plot_slices import plot_center_slices
from postprocessing.utilities.csv_importer import csv_import

filename_rawdat      = '_demo_data/measurement_1/raw_dat.csv'
filename_magnetcoords   = convert_to_magnet_coordinate_frame(filename_rawdat) # relies on centerpoint.csv which is in the same directory as filename_rawdat. For Lakeshore and OSI2-One V1. Other Magnets have other coordinate systems

field, x_values, y_values, z_values = csv_import(filename_magnetcoords)

plot_center_slices(field, x_values, y_values, z_values, title='Center Slices of Bz, Coil Off, in Magnet-Coordinates', unit='mT')
```
